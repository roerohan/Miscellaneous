# 1. Time Series Data

1. Consider the time series data in the attached file “Timeseries data.txt”. Perform thefollowing tasks:
- Visualize the birth rates on monthly basis.
- Visualize the birth rates on even date (2, 4, 6...) 
- Observe the trend of birth on even and odd datesd.Can we predict the birth rate infuture. If so illustrate thesame using appropriate model and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Time series data.txt')
print(df)
```

```python
data = df.to_dict('split')['data']

monthly = {}
for row in data:
    month = (row[0].split('-')[1])
    monthly[month] = monthly.get(month, 0) + row[1]

print(monthly)
```

```python
months = monthly.keys()
births = monthly.values()

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])


ax.bar(months, births)
plt.show()
```

```python
even_dates_data = list(filter(lambda x: int(x[0].split('-')[2]) % 2 == 0, data))
odd_dates_data = list(filter(lambda x: int(x[0].split('-')[2]) % 2 != 0, data))
```

```python
# Birth rate on even dates

fig = plt.figure()

even_dates = list(map(lambda x: int(''.join(x[0].split('-'))), even_dates_data))

even_births = list(map(lambda x: x[1], even_dates_data))

plt.plot(even_dates, even_births, label='even')
plt.xlabel('Dates')
plt.ylabel('Births')
plt.legend()
plt.show()
```

```python
# Birth Rate trend on even dates and odd dates

odd_dates = list(map(lambda x: int(''.join(x[0].split('-'))), odd_dates_data))
odd_births = list(map(lambda x: x[1], odd_dates_data))
plt.plot(even_dates, even_births, label='even')
plt.plot(odd_dates, odd_births, label='odd')
plt.xlabel('Dates')
plt.ylabel('Births')
plt.legend()
plt.show()
```

```python
# Birth Rates can be predicted using Logistic regression Machine Learning model

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(C=1e5)
logreg.fit
Z = logreg.predict


# Visualization is as follows

plt.plot(even_dates[:len(even_dates)//2], even_births[:len(even_births)//2], label='even')
plt.plot(odd_dates[:len(odd_dates)//2], odd_births[:len(odd_births)//2], label='odd')
plt.xlabel('Dates')
plt.ylabel('Births')
plt.legend()
plt.show()
```

```python
# Function to find mean
def mean(l):
    return sum(l)/len(l)

# Configurations array
configs = [
    ('0.25J', '100 X 100'),
    ('0.25J', '200 X 100'),
    ('0.25J', '300 X 100'),
    ('0.5J', '100 X 100'),
    ('0.5J', '200 X 100'),
    ('0.5J', '300 X 100'),
    ('1J', '100 X 100'),
    ('1J', '200 X 100'),
    ('1J', '300 X 100'),
]
```

## Network Stability

```python
x = []

x.append(mean(data['0.25J']['100']['first']))
x.append(mean(data['0.25J']['200']['first']))
x.append(mean(data['0.25J']['300']['first']))
x.append(mean(data['0.5J']['100']['first']))
x.append(mean(data['0.5J']['200']['first']))
x.append(mean(data['0.5J']['300']['first']))
x.append(mean(data['1J']['100']['first']))
x.append(mean(data['1J']['200']['first']))
x.append(mean(data['1J']['300']['first']))

print('List of mean round at first dead')
print(x)

print()

print('The maximum network stability can be obtained by the following configuration:')
res = configs[x.index(max(x))]

print('Energy: ' + res[0])
print('Grid Area: ' + res[1])

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

confs = []

for c in configs:
    confs.append(' '.join(c))

ax.bar(confs, x)
plt.xlabel('Configuration')
plt.ylabel('Round at first dead')

plt.show()
```

## Network Lifetime

```python
x = []
x.append(mean(data['0.25J']['100']['last']))
x.append(mean(data['0.25J']['200']['last']))
x.append(mean(data['0.25J']['300']['last']))
x.append(mean(data['0.5J']['100']['last']))
x.append(mean(data['0.5J']['200']['last']))
x.append(mean(data['0.5J']['300']['last']))
x.append(mean(data['1J']['100']['last']))
x.append(mean(data['1J']['200']['last']))
x.append(mean(data['1J']['300']['last']))

print('List of mean round at all dead:')
print(x)

print()

print('The maximum network lifetime can be obtained by the following configuration:')
res = configs[x.index(max(x))]

print('Energy: ' + res[0])
print('Grid Area: ' + res[1])

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

confs = []

for c in configs:
    confs.append(' '.join(c))

ax.bar(confs, x)
ax.bar(confs, x)
plt.xlabel('Configuration')
plt.ylabel('Round at all dead')

plt.show()
```
